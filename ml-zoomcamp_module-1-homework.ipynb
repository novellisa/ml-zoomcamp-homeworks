```python
lotuses_red = lotuses.drop(columns=['Make', 'Model', 'Year', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Number of Doors', 'Market Category', 'Vehicle Size', 'Vehicle Style', 'highway MPG', 'city mpg', 'Popularity', 'MSRP'])
lotuses_red
```

```python
lotuses_red = lotuses_red.drop_duplicates()
lotuses_red
```

```python
lotuses_red.shape
```

```python
X = lotuses_red.values
X
```

```python
print(X.shape)
XTX = X.T.dot(X)
print(XTX.shape)
```

```python
XTX_inv = np.linalg.inv(XTX)
```

```python
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
```
